```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("dataset/WA_Fn-UseC_-HR-Employee-Attrition.csv")

print("Dataset Loaded Successfully")
print(df.head())
```

```python
print("Shape:", df.shape)

print("\nColumns:")
print(df.columns.tolist())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nData Types:")
print(df.dtypes)
```

```python
df.describe()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Attrition', data=df)
plt.title('Employee Attrition Distribution')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', hue='Attrition', data=df)
plt.title('Attrition by Gender')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(y='Department', hue='Attrition', data=df)
plt.title('Attrition by Department')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['MonthlyIncome'], bins=30)
plt.title('Monthly Income Distribution')
plt.show()
```

```python
plt.figure(figsize=(15,10))

numeric_df = df.select_dtypes(include=np.number)

sns.heatmap(
    numeric_df.corr(),
    cmap='coolwarm'
)

plt.title('Correlation Heatmap')
plt.show()
```

```python
df.shape
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='OverTime', hue='Attrition', data=df)
plt.title('Attrition by Overtime')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(y='JobRole', hue='Attrition', data=df)
plt.title('Attrition by Job Role')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='Age', hue='Attrition', bins=20)
plt.title('Attrition by Age')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title('Monthly Income vs Attrition')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Attrition', data=df)
plt.title('Employee Attrition Distribution')
plt.savefig('attrition_distribution.png')
plt.show()
```

```python
print("Total Employees:", len(df))

print("Employees Left:")
print(df['Attrition'].value_counts())

attrition_rate = (df['Attrition'] == 'Yes').mean() * 100
print("\nAttrition Rate:", round(attrition_rate, 2), "%")
```

```python
df.to_csv("cleaned_hr_attrition.csv", index=False)
print("File Saved")
```
